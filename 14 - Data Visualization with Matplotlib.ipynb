Import the `matplotlib.pyplot` module under the name `plt` (the tidy way):

```python
import matplotlib.pyplot as plt
%matplotlib inline
# This enables us to view our graphs in the notebook
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# To create a simple plot in matplotlib we use the following commands
plt.plot(x,y)
plt.xlabel('Values of X')
plt.ylabel('Values of Y')
plt.title('Line Plot of X against X**2')
```

## Creating Muliple plots on same canvas

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r')
plt.subplot(2, 1, 2)
plt.plot(y, x, 'black')
```

### Matplotlib Object Oriented Method
The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

To begin we create a figure instance. Then we can add axes to that figure:

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
# left, bottom, width, height (range 0 to 1)
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'r')
axes1.set_xlabel("Values of X")
axes1.set_ylabel("X ** 2")
axes1.set_title('Line Plot of X against X**2')

axes2 = fig1.add_axes([0.2, 0.4, 0.3, 0.3])
axes2.plot(y, x, 'black')
axes2.set_xlabel("X ** 2")
axes2.set_ylabel("Values of X")
axes2.set_title('Line Plot of X against X**2')
```

### Figure size, aspect ratio and DPI
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig2 = plt.figure(figsize=(4, 3))
axes3 = fig2.add_axes([0, 0, 1, 1])
axes3.plot(x, y)
```

### Saving figures
Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF.

To save a figure to a file we can use the savefig method in the Figure class:

```python
fig2.savefig('Data Viz.pdf')
fig1.savefig('Data Viz.jpg')
```

### Legends

```python
fig4 = plt.figure(figsize=(4,3))
ax = fig4.add_axes([0,0,1,1])
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.set_title("Values of X against it's cube and square")
ax.legend()
```

### Linewidth, linestyle, marker styles and colors

```python
fig5 = plt.figure(figsize=(10,6))
ax1 = fig5.add_axes([0, 0, 1, 1])

ax1.plot(x, x+1, color='red', linewidth=0.25)
ax1.plot(x, x+2, color='red', linewidth=0.5)
ax1.plot(x, x+3, color='red', linewidth=1)
ax1.plot(x, x+4, color='red', linewidth=2)

# possible linestyle options -, --, -., :
ax1.plot(x, x+5, color='green', lw=2, linestyle='-')
ax1.plot(x, x+6, color='green', lw=2, linestyle='--')
ax1.plot(x, x+7, color='green', lw=2, linestyle='-.')
ax1.plot(x, x+8, color='green', lw=2, linestyle=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax1.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+', markersize=10)
ax1.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax1.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax1.plot(x, x+12, color="blue", lw=3, ls='--', marker='1', markersize=10)
```
